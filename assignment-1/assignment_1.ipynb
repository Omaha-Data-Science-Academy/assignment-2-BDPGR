Plotting a dataset is a great way to explore its distribution. Plotting the iris dataset can be done using matplotlib, a Python library for 2D plotting.

```python
# Plotting in MatPlotLib
import pandas as pd
import matplotlib.pyplot as plt
# Create a scatter plot
plt.figure(figsize=(10, 6))

# Plot each species with a different color
species = df['species'].unique()
colors = ['red', 'green', 'blue']

for species, color in zip(species, colors):
    species_data = df[df['species'] == species]
    plt.scatter(species_data['petal_length'], species_data['petal_width'],
                c=color, label=species, alpha=0.7)

# Customize the plot
plt.title('Iris Dataset - Petal Dimensions', fontsize=16)
plt.xlabel('Petal Length (cm)', fontsize=14)
plt.ylabel('Petal Width (cm)', fontsize=14)
plt.legend()

# Add a grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```

## Models for Data (15 points)

A model is a function that takes an input $x$ and produces a prediction $\hat{y}$.

Let's consider two possible models for this data:
1. $M_1(x) = x+5$
2. $M_2(x) = 2x+1$

Compute the predictions of models $M_1$ and $M_2$ for the values in $X$. These predictions should be vectors of the same shape as $Y$. Then plot the prediction lines of these two models overlayed on the "observed" data $(X, Y)$. Use [plt.plot()](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html) to draw the lines.
```python
import matplotlib.pyplot as plt
import numpy as np

# Example observed data (X and Y)
X = np.array([1, 2, 3, 4, 5])  # Example X values (replace with your actual data)
Y = np.array([7, 8, 10, 12, 14])  # Example Y values (replace with your actual data)

# Compute predictions for M1 and M2
M1 = X + 5  # Model 1: y = x + 5
M2 = 2 * X + 1  # Model 2: y = 2x + 1

# Scatter plot of observed data
plt.scatter(X, Y, color='b', label='Observed Data')

# Plotting the prediction lines for M1 and M2
plt.plot(X, M1, color='r', label='M1(x) = x + 5')
plt.plot(X, M2, color='g', label='M2(x) = 2x + 1')

# Adding legend, title, and labels
plt.legend(loc="upper left")

font1 = {'family': 'serif', 'color': 'blue', 'size': 20}
font2 = {'family': 'serif', 'color': 'darkred', 'size': 15}

plt.title("Comparison of M1 and M2 Models", fontdict=font1)
plt.xlabel("X Input Values", fontdict=font2)
plt.ylabel("Y Output Values", fontdict=font2)

# Show plot
plt.show()

```
